```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
plt.style.use('classic')
mas = [i for i in range(100)]
plt.plot(mas)
```

```python
x = np.linspace(0,10,100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```
