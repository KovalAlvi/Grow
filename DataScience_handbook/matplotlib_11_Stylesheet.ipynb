```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
plt.style.available
```

```python
plt.style.use('bmh')
x = np.random.randn(1000)
plt.hist(x)
```

```python
def hist_and_lines():
    np.random.seed(2)
    fig, ax  =plt.subplots(1,2, figsize=(11,4))
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a','b', 'c'], loc='lower left')
   
```

```python
for name_style in plt.style.available:
    with plt.style.context(name_style):
        hist_and_lines()
        plt.title(name_style, color = 'purple')
        
        
    
```
