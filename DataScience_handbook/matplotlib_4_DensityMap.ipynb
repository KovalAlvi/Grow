```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("seaborn-whitegrid")
import numpy as np
```

```python
### Визуализация трех мерных данных на двумерной плоскости
### plt.contor plt.contorf plt.imshow
```

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10+x*y)*np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y) # для создания двумерной сетки, для дальнейшего изменения
Z = f(X, Y)
```

```python
print(x, y, sep='\n')
```

```python
plt.contour(X, Y, Z, color='black')
```

```python
plt.contour(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
```

```python
plt.imshow(Z, extent=[0,5,0,5], origin='lower', cmap='RdGy') #origin = lower мы переварачиваем
plt.colorbar()
plt.axis(aspect='image')
```

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=12)
plt.imshow(Z, extent=[0,5,0,5], origin='lower', cmap='RdGy', alpha=0.5) #origin = lower мы переварачиваем
plt.colorbar()
```
