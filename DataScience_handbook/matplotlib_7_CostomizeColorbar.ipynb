```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("classic")
import numpy as np
```

```python
x = np.linspace(1,10,10)
print(x)
I = (x) * x[:, np.newaxis]
print(np.sin(x).shape,np.cos(x[:, np.newaxis]).shape)
print(I, I.shape)
```

```python
x = np.linspace(1,10,1000)
I = np.sin(x) * np.cos(x[:, np.newaxis])
plt.imshow(I)
plt.colorbar()
```

```python
x = np.linspace(1,10,1000)
y = np.linspace(1,10,1000)
X, Y = np.meshgrid(x, y)
I = np.sin(X) * np.cos(Y)
plt.imshow(I)
plt.colorbar()
```

```python
plt.imshow(I, cmap='cubehelix')
plt.colorbar()
```

```python
speckles = (np.random.random(I.shape) <0.01)
I[speckles] = np.random.normal(0,3, np.count_nonzero(speckles))

print(speckles.sum(), I.shape)
plt.figure(figsize=(10,3.5))

plt.subplot(1,2,1)
plt.imshow(I, cmap='RdBu')
plt.colorbar()

plt.subplot(1,2,2)
plt.imshow(I, cmap='RdBu')
plt.colorbar(extend='both')
plt.clim(-1,1)
```

```python
#Мы можем делать и дискретное отображение непрерыной величины
plt.imshow(I, cmap=plt.cm.get_cmap('Blues', 6))
plt.clim(-1,1)
plt.colorbar()
```
