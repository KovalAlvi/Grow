## plt.axes Subplots by hand

```python
ax1 = plt.axes()
ax2 = plt.axes([0.65, 0.65, 0.2, 0.2])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4],
                  xticklabels=[], ylim = (-1.2, 1.2))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4],
                  ylim = (-1.2, 1.2))
x = np.linspace(0,10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x))
```

## plt.subplot Simple Grid

```python
for i in range(1,7):
    plt.subplot(2,3, i)
    plt.xticks([])
    plt.subplots_adjust(hspace=0.1, wspace=0.2) # расстояние между графиками
    plt.text(0.5, 0.5, str((2,3,i)),
            fontsize=18, ha='center')
    
    
```

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1,7):
    ax = fig.add_subplot(2,3,i)
    ax.text(0.5, 0.5, str((2,3,i)),
            fontsize=18, ha='center')
    ax.set(xticks=[])
```

## plt.subplots : The whole Grid in One Go


```python
fig, ax = plt.subplots(2, 3,  sharex='col', sharey='row')
```

```python
for i in range(2):
    for j in range(3):
        ax[i,j].text(0.5, 0.5, str((i,j)),
            fontsize=18, ha='center')
        
fig
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2, 3,  sharex='col', sharey='row')
for i , axi in enumerate(ax.flat):
    axi.text(0.5, 0.5, str((2,3, i+1)),
            fontsize=18, ha='center')
```

```python
## plt.GridSpec : More complicated
```

```python
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)
plt.subplot(grid[0,0])
plt.subplot(grid[0,1])
plt.subplot(grid[:,2])
plt.subplot(grid[1,0])
plt.subplot(grid[1,1])
```

```python
mean = [0,0]
cov = [[1,2],
       [2,2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

fig = plt.figure(figsize=(6,6))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey = main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex = main_ax)

#scatter points on the main axes
main_ax.plot(x, y, 'ob', markersize=8, alpha=0.01)

#histogram on the attached
x_hist.hist(x, 40, histtype='stepfilled',
      orientation='vertical', color='purple', alpha=0.2)
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype='stepfilled',
      orientation='horizontal', color='b',  alpha=0.2)
y_hist.invert_xaxis()

```
