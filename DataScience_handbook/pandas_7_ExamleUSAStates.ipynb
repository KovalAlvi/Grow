```python
pop = pd.read_csv('state-population.csv')
areas = pd.read_csv('state-areas.csv')
abbrevs = pd.read_csv('state-abbrevs.csv')
print(pop.head(), areas.head(), abbrevs.head(), sep = '\n\n')
```

```python
merged = pd.merge(pop, abbrevs, how='outer',
                 left_on='state/region', right_on='abbreviation')
merged = merged.drop('abbreviation', axis =1)
merged
```

```python
merged.isnull().any()
```

```python
merged[merged['population'].isnull()]
```

```python
merged.loc[merged['state'].isnull()]['state/region'].unique()
```

```python
merged.loc[merged['state/region'] == 'PR', 'state'] = 'Puerto Rico'
merged.loc[merged['state/region'] == 'USA', 'state'] = 'United States'
```

```python
merged.isnull().any()
```

```python
final = pd.merge(merged, areas, on='state', how='left')
final.head()
```

```python
final.isnull().any()
```

```python
final['state'][final['area (sq. mi)'].isnull()].unique()
```

```python
final.dropna(inplace=True)
print(final.isnull().any())
final
```

```python
#Есть несколько способов провести фильтрацию данных. Прямо через маску или через query
mask = (final['year']==2010) & (final['ages'] == 'total')
data2010 = final[mask]
data2010.head()
```

```python

data2010 = final.query("year==2010 & ages == 'total'")
data2010.head()
```

```python
data2010.set_index('state', inplace=True)
density = data2010['population'] / data2010['area (sq. mi)']
```

```python
density.sort_values(ascending=False, inplace=True) #Очень важно делать inplace= TRUE чтобы не создавать новое
density.head()
```
