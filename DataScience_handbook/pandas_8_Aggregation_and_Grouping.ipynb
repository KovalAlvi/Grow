## GroupBy: Split, Apply. Combine

```python
df = pd.DataFrame({'key': list('ABCABC'),
                  'data': range(6)}, columns = ['key', 'data'])
df
```

```python
df.groupby('key').sum()
```

```python
planets.groupby('method')['orbital_period'].median()
```

```python
planets.groupby('method')['year'].describe()
```

```python
rng = np.random.RandomState(0)
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data1': range(6),
                   'data2': rng.randint(0, 10, 6)},
                   columns = ['key', 'data1', 'data2'])
df
```

```python
### Aggregate
```

```python
df.groupby('key').aggregate([np.min, np.median, np.max])
```

```python
### Filtering
```

```python
df
```

```python
#он объединяет в группы по keys затем к ним применяет функции какие-то в отдельности
df.groupby('key').transform(lambda x: x-x.mean())
```

```python
# Хороший пример по структуризации данных. Мы показываем чт было 00 или 10 или 90 и оставляем индекс
# Группируем по методу и по годам 
# действительно очень удобно

decade = 10* (planets['year'] //10 )
decade = decade.astype(str) + 's'
decade.name = 'decade'
planets.groupby(['method', decade])['number'].sum().unstack().fillna(0)
```
